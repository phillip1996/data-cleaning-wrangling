```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing file
df =pd.read_excel('HousingData.xlsx')
df.head(5)
```

```python
# dropping irrelevant columns
columns = ['TaxDistrict','LegalReference']
df.drop(columns, axis = 1, inplace =True )
```

```python
# dropping irrelevant rows
rows = [159,223,246,531,533,665,858,1043,1176,1262,1642,1713,1714,1782,1921,1210,2262,3434,3681]
df.drop(rows, axis = 0, inplace =True )
```

```python
# checking for duplicate entries
df.duplicated().sum()
```

```python
# checking values ranging from 0 to 534 toconfirm removal of rows specifically in the PropertyAddress column
df.head(534)
```

```python
# replacing missing entry with owners name
df[['OwnerName']] = df[['OwnerName']].fillna('TONY,REX DASHA')
```

```python
# replacing missing entry with owners Address
df[['OwnerAddress']] = df[['OwnerAddress']].fillna('300 BENTON CT, GOODLETTSVILLE, TN')
```

```python
# checking if owners name & address has been inplaced
df.head(20)
```

```python
# replacing N/A values with zero
df = df.fillna(0)
```

```python
# re-checking for missing value
df.isnull().sum()
```

```python
# renaming column
df.rename(columns=({ 'SaleDate': 'DateOfSales'}), inplace=True)
```
